```python
import numpy as np
import healpy as hp
import pymaster as nmt
```

```python
scale = 10

# Intersection masks

int_baseline = hp.read_map('../../BmodeLike/masks/intersection_ns0064_baseline_mask.fits')
int_baseline = nmt.mask_apodization(int_baseline, scale, apotype='C2')
fsky = np.mean(int_baseline**2)
hp.mollview(int_baseline, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_baseline_nside64_aposcaleTest%s.npy' % scale, int_baseline, overwrite=True)

int_medium = hp.read_map('../../BmodeLike/masks/intersection_ns0064_medium_mask.fits')
int_medium = nmt.mask_apodization(int_medium, scale, apotype='C2')
fsky = np.mean(int_medium**2)
hp.mollview(int_medium, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_medium_complexity_nside64_aposcaleTest%s.npy' % scale, int_medium, overwrite=True)

int_high = hp.read_map('../../BmodeLike/masks/intersection_ns0064_high_mask.fits')
int_high = nmt.mask_apodization(int_high, scale, apotype='C2')
fsky = np.mean(int_high**2)
hp.mollview(int_high, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_high_complexity_nside64_aposcaleTest%s.npy' % scale, int_high, overwrite=True)
```

```python
def smooth_mask(mask, sigma=3, threshold=1e-2):
    mask = nmt.mask_apodization(mask, sigma, apotype='Smooth')
    mask[mask<threshold], mask[mask>threshold] = 0, 1
    mask[mask==0] = 0
    
    return mask
```

```python
scale = 5

mask0 = hp.read_map('../../BmodeLike/masks/intersection_ns0064_baseline_mask.fits')
hp.mollview(mask0, title=r'$f_{\rm sky} = %s$' % (np.mean(mask0**2)))

mask = smooth_mask(mask0, sigma=3, threshold=1e-2)
hp.mollview(mask, title=r'$f_{\rm sky} = %s$' % (np.mean(mask**2)))

comp = np.copy(mask)
comp[mask0==0] = -1
hp.mollview(comp, title='Comparison')

mask = nmt.mask_apodization(mask, scale, 'C2')
hp.mollview(mask, title=r'$f_{\rm sky} = %s$' % (np.mean(mask**2)))
```

```python
scale = 5

# Intersection masks

int_baseline = hp.read_map('../../BmodeLike/masks/intersection_ns0064_baseline_mask.fits')
int_baseline = smooth_mask(int_baseline)
int_baseline = nmt.mask_apodization(int_baseline, scale, apotype='C2')
fsky = np.mean(int_baseline**2)
hp.mollview(int_baseline, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_baseline_nside64_aposcale%s.npy' % scale, int_baseline, overwrite=True)

int_medium = hp.read_map('../../BmodeLike/masks/intersection_ns0064_medium_mask.fits')
int_medium = smooth_mask(int_medium)
int_medium = nmt.mask_apodization(int_medium, scale, apotype='C2')
fsky = np.mean(int_medium**2)
hp.mollview(int_medium, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_medium_complexity_nside64_aposcale%s.npy' % scale, int_medium, overwrite=True)

int_high = hp.read_map('../../BmodeLike/masks/intersection_ns0064_high_mask.fits')
int_high = smooth_mask(int_high)
int_high = nmt.mask_apodization(int_high, scale, apotype='C2')
fsky = np.mean(int_high**2)
hp.mollview(int_high, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_high_complexity_nside64_aposcale%s.npy' % scale, int_high, overwrite=True)
```
