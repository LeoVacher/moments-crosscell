```python
import sys
sys.path.append('../lib/')
import numpy as np
import healpy as hp
import pymaster as nmt
import matplotlib.pyplot as plt
from tqdm import trange
from mpfit import mpfit
import simu_lib as sim
```

```python
# Inputs

Nside = 64
lmax = 2*Nside-1
Nlbin = 1
Nsims = 250
masking_strat = '_fsky0.7'
scale = 10

Npixs = hp.nside2npix(Nside)
```

```python
# Load instrument

instr_name = 'litebird_full'
instr =  np.load("../lib/instr_dict/%s.npy"%instr_name,allow_pickle=True).item()

freq = instr['frequencies']
sens = instr['sens_P']
beam = instr['beams']

Nfreqs = len(freq)
Ncross = int(Nfreqs*(Nfreqs+1) / 2)

sigma = sens / hp.nside2resol(Nside, arcmin=True)
```

```python
# Load and bin CL_noise

b = nmt.NmtBin.from_lmax_linear(lmax, nlb=Nlbin, is_Dell=False)
leff = b.get_effective_ells()
Nbins = len(leff)

CL_noise = np.load('./CL_noise%s_nside%s_aposcale%s.npy' % (masking_strat, Nside, scale))[:Nsims]
CL_noise_binned = np.zeros((Nsims, Nfreqs, 2, Nbins))

for k in range(Nsims):
    for i in range(Nfreqs):
        CL_noise_binned[k, i] = b.bin_cell(CL_noise[k, i, 1:, :2*Nside])

CL_noise_mean = np.mean(CL_noise_binned, axis=(0,2))
CL_noise_std = np.std(CL_noise_binned, axis=(0,2))
```

```python
# Fit noise model

def N(ell, N_white, N_corr, l_knee, alpha_knee):
    return N_white + N_corr * (ell/l_knee)**alpha_knee

def chi2(p, fjac=None, ell=None, data=None, sigma=None):
    N_white, N_corr, l_knee, alpha_knee = p
    model = N(ell, N_white, N_corr, l_knee, alpha_knee)

    status = 0
    return [status, (data - model)**2 / sigma**2]

Nparams = 4
N_white = np.zeros(Nfreqs)
N_corr = np.zeros(Nfreqs)
l_knee = np.zeros(Nfreqs)
alpha_knee = np.zeros(Nfreqs)

for i in range(Nfreqs):
    p = []
    p.append({'value': 4*np.pi/Npixs * sigma[i]**2, 'fixed': 0, 'limited': [1,1], 'limits': [0,np.inf]}) # N_white
    p.append({'value': 0, 'fixed': 1, 'limited': [1,1], 'limits': [0,np.inf]}) # N_corr
    p.append({'value': 100, 'fixed': 1, 'limited': [1,1], 'limits': [1,np.inf]}) # l_knee
    p.append({'value': -1, 'fixed': 1, 'limited': [1,1], 'limits': [-10,0]}) # alpha_knee

    fa = {'ell': leff, 'data': CL_noise_mean[i], 'sigma': CL_noise_std[i]}
    m = mpfit(chi2, parinfo=p, functkw=fa, quiet=1)

    N_white[i], N_corr[i], l_knee[i], alpha_knee[i] = m.params
```

```python
ell = np.geomspace(2, 2*Nside-1, 100)

for i in range(Nfreqs):
    plt.errorbar(leff, CL_noise_mean[i], yerr=CL_noise_std[i], fmt='o', color='black', alpha=0.6)
    plt.loglog(leff, 4*np.pi/Npixs * sigma[i]**2 * np.ones(Nbins), linestyle=(0, (10,6)), linewidth=1, color='black')
    plt.loglog(ell, N(ell, N_white[i], N_corr[i], l_knee[i], alpha_knee[i]), linewidth=2, color='red', zorder=3)
    plt.xlabel(r'$\ell$')
    plt.ylabel(r'$\mathcal{C}_\ell^{\rm noise}(\nu)$ [$\mu$K$^2$]')
    plt.title(r'$\nu =$%s GHz' % (freq[i]))
    plt.show()
```
