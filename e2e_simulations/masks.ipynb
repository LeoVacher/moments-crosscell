```python
import numpy as np
import healpy as hp
import pymaster as nmt
```

```python
scale = 3

# Intersection masks

int_baseline = hp.read_map('../../BmodeLike/masks/intersection_ns0064_baseline_mask.fits')
int_baseline = nmt.mask_apodization(int_baseline, scale, apotype='C2')
fsky = np.mean(int_baseline**2)
hp.mollview(int_baseline, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_baseline_nside64_aposcale%s.npy' % scale, int_baseline, overwrite=True)

int_medium = hp.read_map('../../BmodeLike/masks/intersection_ns0064_medium_mask.fits')
int_medium = nmt.mask_apodization(int_medium, scale, apotype='C2')
fsky = np.mean(int_medium**2)
hp.mollview(int_medium, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_medium_complexity_nside64_aposcale%s.npy' % scale, int_medium, overwrite=True)

int_high = hp.read_map('../../BmodeLike/masks/intersection_ns0064_high_mask.fits')
int_high = nmt.mask_apodization(int_high, scale, apotype='C2')
fsky = np.mean(int_high**2)
hp.mollview(int_high, title=r'$f_{\rm sky} = %s$' % (fsky))
hp.write_map('../masks/mask_intersection_high_complexity_nside64_aposcale%s.npy' % scale, int_high, overwrite=True)
```
